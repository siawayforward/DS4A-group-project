```python
#Code for creating dictionaries of Bay Area ZipCodes from 2010 Census Data

import pandas as pd
import json
import numpy as np

#Importing Clean Info of Bay Area Counties
BA_county = pd.read_csv("data/BAY_AREA_COUNTY_2010.csv")
#renaming
BA_county.index = BA_county["NAME"]

#The first two digits in the COunty ID are the state ID (5-Digit Number)

BA_county
```

```python
#Importing US Zip Codes

US_zip = pd.read_csv("data/US_ZIP_CODE_2010.csv")
US_zip.index = US_zip["GEOID"]
US_zip = US_zip.rename(columns={"GEOID": "ZIP_CODE"})

US_zip.head(5)
```

```python
#Importing US ZIP to County Data

zip_county = pd.read_csv("data/zip_county_2010.csv")

# Removing deplicates by keeping the one with the largest area
zip_county = zip_county.sort_values(by = ["GEOID","AREALANDPT"],ascending = False).drop_duplicates().reset_index()
zip_county.head(15)
```

```python
#Creating Bay Area JSON File
BA_county_j = BA_county.to_dict(orient="index")
BA_county_j
```

```python
county_list = list(BA_county["NAME"])
all_ba_zc = [];
US_zip["COUNTY"] = ""
for county in county_list:
    c_zc = list(zip_county.loc[zip_county['GEOID'] == BA_county_j[county]["GEOID"]]["ZCTA5"])
    US_zip.loc[c_zc,"COUNTY"] = county;
    #Starting Loop to Clean Data and Create Bay Area County JSON
    BA_county_j[county].update({"ZIP_CODES":list(c_zc)})
    all_ba_zc.extend(list(c_zc))
all_ba_zc = list(set(all_ba_zc))
all_ba_zc
```

```python
BA_county_j
```

```python
BA_zip_codes = US_zip.loc[all_ba_zc,:].sort_values(by=["ZIP_CODE"])

#Saving Bay Area Zip Codes into csv and json file

BA_zip_codes.to_csv('Data/BA_ZIP_CODES.csv', index=False)

with open('Data/BA_ZIP_CODES.json', 'w', encoding='utf-8') as f:
    json.dump(BA_zip_codes.to_dict('index'), f, ensure_ascii=False, indent=4)

BA_zip_codes
```
